```python
import seaborn as sns
import pandas as pd
import scipy as sp
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from sklearn.manifold import TSNE
from sklearn.preprocessing import MinMaxScaler
from scipy.stats import multivariate_normal
from sklearn.mixture import GaussianMixture

%matplotlib inline

n_cluster = 6
data_dim = 50
num_samples = 400

set_mu = []
set_cov = []

for i in range(1, n_cluster+1):
    mean = np.zeros(data_dim) + i * i
    for j in range(data_dim) :
        mu_j = np.random.choice([-1, 1]) * mean[j]
        mean[j] = mu_j
    set_mu.append(mean)

# 다변량 정규분포 생성
toydata = []
for i in range(n_cluster):
    covariance_matrix = np.diag(np.ones(data_dim))
    mvn = multivariate_normal(mean=set_mu[i], cov=covariance_matrix)
    sample = mvn.rvs(num_samples)
    for s in sample:
        toydata.append(s)

scaler = MinMaxScaler()
X = toydata
print(X)
toydata = pd.DataFrame(toydata)
scaler.fit(toydata)
toydata = scaler.transform(toydata)
toydata = pd.DataFrame(toydata)
class CustomDataset(Dataset):
    def __init__(self, csv):
        self.inp = csv.values
        lables = []
        for i in range(1, 7) :
            for j in range(400):
                lables.append(i)
        lables = pd.DataFrame(lables)
        self.lable = lables.values

    def __len__(self):
        return len(self.inp) 

    def __getitem__(self,idx):
        inp = torch.FloatTensor(self.inp[idx])
        lable = torch.LongTensor(self.lable[idx])
        return inp, lable
    
dataset = CustomDataset(toydata)
dataloader = DataLoader(dataset, batch_size=5, shuffle=True)


tsne = TSNE(n_components=2, random_state=50)
X_tsne = tsne.fit_transform(toydata)


n_components = 6  # 클러스터 수 (가우시안 분포의 수)
gmm = GaussianMixture(n_components=n_components)
gmm.fit(X)
labels = gmm.predict(X)

# 클러스터 중심 및 클러스터 색상 설정
cluster_centers = gmm.means_
colors = ['r', 'b']

# 데이터 포인트 시각화
plt.figure(figsize=(8, 6))
for i in range(n_components):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], c=colors[i])

# 클러스터 중심 시각화
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='black', marker='x', s=100, label='Cluster Centers')

plt.title('GMM 클러스터링 결과')
plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
