```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from torchvision import datasets, transforms

CSV_PATH = './stats2021.csv'
scaler = MinMaxScaler()
mlb = pd.read_csv(CSV_PATH)
mlb = mlb.iloc[0:, 4:15]
scaler.fit(mlb)
norm_mlb = scaler.transform(mlb)
norm_mlb = pd.DataFrame(norm_mlb)

class CustomDataset(Dataset):
	def __init__(self, data):
		self.inp = data
	
	def __len__(self):
		return len(self.inp) 

	def __getitem__(self,idx):
		inp = torch.FloatTensor(self.inp[idx])
		return inp 
    
dataset = CustomDataset(norm_mlb)
dataloader = DataLoader(dataset, batch_size=10, shuffle=True)

data_set = datasets.MNIST('./data', train=True, download=True,
                             transform=transforms.ToTensor())
data_loader = torch.utils.data.DataLoader(
    data_set, batch_size=3, shuffle=True)

for batch in dataloader :
    print(batch)
    
```
